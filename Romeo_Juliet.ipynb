```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x_arrival = []
y_arrival = []
sample_size = 100000
```

```python
x_arrival = np.random.randint(0, 3600, sample_size)
y_arrival = np.random.randint(0, 3600, sample_size)
arrival = pd.DataFrame({'x': x_arrival, 'y': y_arrival, 'delay_time': [15*60]*sample_size})
arrival
```

```python
meets = 0
for i in range(sample_size):
    if arrival['x'][i] > arrival['y'][i]:
        if arrival['x'][i] > arrival['y'][i] + arrival['delay_time'][i]:
            continue
        else:
            meets +=1
    if arrival['y'][i] > arrival['x'][i]:
        if arrival['y'][i] > arrival['x'][i] + arrival['delay_time'][i]:
            continue
        else:
            meets +=1
    if arrival['y'][i] == arrival['x'][i]:
        meets +=1
    else:
        continue
```

```python
print(meets)
print("The probability they meet is: ",meets/sample_size)
print("The theoretical probability they meet is: ",7/16)
```

```python
plt.figure(figsize=(8, 5))
plt.plot([-0.25,1], [0,1.25],'k')
plt.plot([0,1.25], [-0.25,1],'k')
x_fill = [0, -0.25, 1, 1.25]
y_fill = [-0.25, 0, 1.25, 1]
plt.fill(x_fill, y_fill, color='#89cff0', alpha=0.5, label='Meeting region')
plt.title('The meeting of Romeo and Juliet')
plt.xlabel('Arrive of Romeo')
plt.ylabel('Arrive of Juliet')
plt.xlim(0,1)
plt.ylim(0,1)
plt.xticks(np.arange(0, 1.1, step=0.25))
plt.yticks(np.arange(0, 1.1, step=0.25))
plt.legend()
plt.grid()
plt.savefig('RnJplot.png',dpi=100)
plt.show()
```

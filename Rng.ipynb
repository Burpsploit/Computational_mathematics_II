```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
lambda_ = 2
sample_size = 100000
u = np.random.uniform(0,1,sample_size)
```

```python
x = np.log(1-u)/(-lambda_)
```

```python
x_numpy = np.random.exponential(scale=1/lambda_,size=sample_size)
```

```python
plt.figure(figsize=(11,5))
plt.subplot(1,2,1)
plt.hist(x, bins=100, density=True, alpha=0.9, label='Sampled',color='#9FE2BF')
x_vals = np.linspace(0, np.max(x), 1000)
pdf = lambda_ * np.exp(-lambda_ * x_vals)
plt.plot(x_vals, pdf,color='#9F2B68',linewidth=2, label=r'$\lambda \cdot e^{-\lambda  x}$ (PDF)')
plt.xticks(np.arange(-1, 4, step=5))
plt.xlabel('Bins')
plt.ylabel('Density')
plt.title("Exponential random number generator")
plt.xlim(0,2)
plt.legend()
plt.subplot(1,2,2)
plt.hist(x_numpy, bins=100, density=True, alpha=0.9, label='numpy sampled',color='#CC7722')
x_vals = np.linspace(0, np.max(x), 1000)
pdf = lambda_ * np.exp(-lambda_ * x_vals)
plt.plot(x_vals, pdf, color='#654321',linewidth=2, label=r'$\lambda \cdot e^{-\lambda  x}$ (PDF)')
plt.xticks(np.arange(-1, 4, step=5))
plt.xlabel('Bins')
plt.title("Numpy exponential function")
plt.xlim(0,2)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(11,5))
plt.subplot(1,2,1)
stats.probplot(x, dist="expon", sparams=(0,1/lambda_), plot=plt)
ax = plt.gca()
lines = ax.get_lines()
lines[0].set_color('#FFC300')
lines[1].set_color('k')
plt.title("Q-Q plot of custom exponential rng")
plt.xlabel("Theoretical quantiles")
plt.ylabel("Sample Quantiles")
plt.grid(True)
plt.subplot(1,2,2)
stats.probplot(x_numpy, dist="expon", sparams=(0,1/lambda_), plot=plt)
ax = plt.gca()
lines = ax.get_lines()
lines[0].set_color('#BF40BF')
lines[1].set_color('k')
plt.title("Q-Q plot of numpy's exponential function")
plt.xlabel("Theoretical quantiles")
plt.ylabel("")
plt.grid(True)
plt.tight_layout()
plt.show()
```
